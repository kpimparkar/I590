#### [3) Distribution plots](#DISTRIBUTION-PLOTS)
1] distplot() - univariate plot : takes just one variable as input and shows the distribution of that variable
`sns.distplot(tips['total_bill'],kde=False,bins=40)`
kde = kernel density estimation (line going through all bars)  

2] jointplot() - combines two distribution plots  
`sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind='scatter')  
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind='hex')  
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind='reg')  
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind='resid')  
sns.jointplot(x=tips['total_bill'],y=tips['tip'],data=tips, kind='KDE')`    

3] pairplot() - plots joiont plot for all numerical columns of the data set. param "hue" can be used for categorical columns.  
`sns.pairplot(tips)  
sns.pairplot(tips, hue='sex', palette='coolwarm')`  

4] rugplot() - Plot datapoints in an array as sticks on an axis. Univariate plot.  
`sns.rugplot(tips['tip'])`  

5] kdeplot() - Plots the kernel density estimation curve  
`sns.kdeplot(tips['tip'])`

#### [4) CATEGORICAL PLOTS](#CATEGORICAL-PLOTS)
1] barplot() - x axis is a categorical column and y axis is mean() by default. 'estimator' argument acceots a function to calculate values of y axis   
`sns.barplot(x='sex',y='tip',data=tips)`

2] countplot() - same as barplot but y axis has the count  
`sns.countplot(x='sex', data=tips, hue='day')`

3] boxplot() - plots box and whisker plot  
`sns.boxplot(x='day',y='tip', data=tips, hue='sex')`

4] violinplot() - similar to box plot. looks pretty though.  
`sns.violinplot(x='day',y='tip',data=tips, hue='sex', split=True)    
sns.violinplot(x='day',y='tip',data=tips)`

5] stripplot() - scatter plot for given category  
`sns.stripplot(x='day',y='tip',data=tips, hue='sex')`  

6] swarmplot() - it is similar to violin plot but this one plots each data point. Note-points don't scale well with increased data. Not so good to explain data to executives.  
`sns.swarmplot(x='day',y='tip',data=tips)`

7] catplot() - previously names as factorplot(). This one is a generic function which accepts an argument "kind" with plot type as the value and creates that plot with given data.  
`sns.catplot(x='day', y='total_bill', data=tips, kind='box') <-- creates box plot  
sns.catplot(x='day', y='total_bill', data=tips, kind='violin') <-- creates violin plot  
sns.catplot(x='day', y='total_bill', data=tips, kind='boxen') <-- creates boxen plot  `

#### [5) Matrix plots](#Matrix-plots)
- The data has to be in matrix form.   

1] Heatmap() :  
`sns.heatmap(mat,annot=True, linewidths=1, cmap='magma', linecolor='blue' )`  

2] Clustermaps() : it is a heatmap with simalar elements clustered together.  
`sns.clustermap(mat)`  

#### [6) Grid plots](#Grid-plots)
1) PairGrid() : This plots pairplot but gives wide controls  
`grid_obj = sns.PairGrid(iris)   # Gives an empty grid to plot all numerical columns against each other  
 grid_obj.map_diag(sns.distplot) # plots distribution plot diagonally  
 grid_obj.map_upper(plt.scatter) # upper plots   
 grid_obj.map_lower(sns.kdeplot) # lower plots`  
 
2) FacetGrid()  
`g = sns.FacetGrid(tips,col='time',row='sex')  
g.map(plt.scatter,'total_bill','tip')` 

#### [7) Regression plot](#Regression-plot)
`sns.lmplot()`

# Matrix plots
- The data has to be in matrix form.   

1] Heatmap() :  
`sns.heatmap(mat,annot=True, linewidths=1, cmap='magma', linecolor='blue' )`  

2] Clustermaps() : it is a heatmap with simalar elements clustered together.  
`sns.clustermap(mat)`  

```python
tips.head()
## corr() to calculate the correlation matrix and then creating a heatmap of the same
mat = tips.corr()
sns.heatmap(mat,annot=True, linewidths=1, cmap='magma', linecolor='blue' )
```

```python
flights = sns.load_dataset('flights')
# flights.head()
## pivot_table() to create a pivot table of count of passengers in given month of the given year
mat = flights.pivot_table(columns='year',index='month',values='passengers')
sns.heatmap(mat, cmap='coolwarm')
```

```python
# Cluster map of the pivot table.
# It clusters similar datapoints together
sns.clustermap(mat)
```

```python
iris = sns.load_dataset('iris')
# iris.head()
grid_obj = sns.PairGrid(iris)   # Gives an empty grid to plot all numerical columns against each other
grid_obj.map_diag(sns.distplot) # plots distribution plot diagonally
grid_obj.map_upper(plt.scatter) # upper plots 
grid_obj.map_lower(sns.kdeplot) # lower plots
```

```python
g = sns.FacetGrid(tips,col='time',row='sex')
g.map(plt.scatter,'total_bill','tip')
```
